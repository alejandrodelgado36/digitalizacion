# Cálculo del área de un triángulo
Usando la fórmula de Herón, calculamos el área de cualquier triángulo conociendo la longitud de sus tres lados. Debes escribir un programa que te pida los tres lados y que calcule el área.


> Cálculo del semiperímetro: $s = \frac{a+b+c}{2}$

> Cálculo del área: $área = \sqrt{s(s-a)(s-b)(s-c)}$

```python
#Pedimos el valor entero de cada uno de los lados
ladoA = int(input('Dime el primer lado: '))
ladoB = int(input('Dime el segundo lado: '))
ladoC = int(input('Dime el tercer lado: '))

#Cálculo del semiperímetro
semi = ((ladoA + ladoB + ladoC) / 2)
#Cálculo del área
area = ((semi * (semi - ladoA) * (semi - ladoB) * (semi - ladoC))** 0.5)
#Imprimimos el resultado
print('El semiperímetro es',semi,'y el área es',area)
```
