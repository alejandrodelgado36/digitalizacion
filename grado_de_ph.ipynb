```python
##Recibimos el grado del ph de parte del usuario:
pH = int(input('Dime el grado del pH: '))
if 0<= pH <= 4:
  print('Es un pH muy ácido.')
elif 4 < pH < 7:
  print('Es un pH poco ácido.')
elif ph == 7:
  print('Es un pH neutro.')
elif 7 < pH <= 10:
  print('Es un pH poco alcalino.')
elif 10 < pH <= 14:
  print('Es un pH muy alcalino.')
else:
  print('Ese pH no existe.')
```
